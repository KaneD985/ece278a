```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from lib.img_load import *
from lib.img_seg import *
from lib.img_bin import *
from skimage.segmentation import slic
from skimage import color
from skimage.segmentation import felzenszwalb

# segmented_img = img_seg(img, 30, 40)
# plt.imshow(img, cmap='gray')
# plt.show()
# plt.imshow(segmented_img,cmap='gray')
# plt.show()
from skimage import data, segmentation, color, measure
from skimage.future import graph
from matplotlib import pyplot as plt


img = img_load('data/image_0020.jpg')

labels1 = segmentation.slic(img, compactness=0.1, n_segments=40, sigma = 1, convert2lab=True)

out1 = color.label2rgb(labels1, img, kind='avg', bg_label=40)

g = graph.rag_mean_color(img, labels1)
labels2 = graph.cut_threshold(labels1, g, 1)
out2 = color.label2rgb(labels2, img, kind='avg', bg_label=1)
boundries = find_boundaries(labels2)


fig, ax = plt.subplots(nrows=3, sharex=True, sharey=True,
                       figsize=(6, 8))

ax[0].imshow(img)
ax[1].imshow(out1)
ax[2].imshow(boundries)

plt.tight_layout()
```

```python

print(np.unique(labels2))
segment_num = len(np.unique(labels2))
print(str(segment_num))

print(np.shape(labels2))

a = measure.regionprops(labels2)
for p in range(len(a)):
    print(a[p].centroid)
cord = a[5].coords
print(cord[0])
print(img)


```

```python
def extract_region(img, region_coords):
    region = np.zeros_like(img)
    for coord in region_coords:
        row = coord[0]
        col = coord[1]
        region[row, col, :] = img[row, col, :]
    return region

plate_area = extract_region(img, cord)
plt.imshow(plate_area, cmap='gray')
plt.show()


```

```python
image = plate_area
# Apply adaptive Equalization
image_adapteq = exposure.equalize_adapthist(image, clip_limit=0.1)
# Show original and result
plt.imshow(image_adapteq, cmap='gray')
```

### Contour Detection

```python
import numpy as np
import cv2 as cv
im = plate_area
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

for contour in contours:
    (x,y,w,h) = cv.boundingRect(contour)
    rat = w/h
    if abs(rat - 2) <= 0.1 and w >= 50 and w < = 150:
        cv.rectangle(im, (x,y), (x+w,y+h), (0,255,0), 2)
        
# cv.drawContours(im, contours, -1, (0,255,0), 3)
plt.imshow(im, cmap='gray')
```

### Local Thresholding

```python
from skimage.filters import threshold_local
img = img_load('data/image_0020.jpg')
img = color.rgb2gray(img)
block_size = 77
loc_thres = threshold_local(img, block_size)
loc_bin = img > loc_thres
plt.imshow(loc_bin, cmap='gray')
```
